```python
import os
import cv2
import numpy as np
from PIL import Image
import glob

def load_images_from_folder(folder, nmax=10000):
    images = []
    images_transposed = []
    for filename in os.listdir(folder)[:nmax]:
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
            images_transposed.append(img.transpose((-1, 0, 1)))
    return np.asarray(images), np.asarray(images_transposed)
```

```python
target_directories = glob.glob("../data_to_score/*/")
scores = []
logfile = open("logfile.txt","a+")


for directory in target_directories:
    print(directory)

    try:
        images, images_transposed = load_images_from_folder(directory)
        current_score, current_std = get_inception_score(images_transposed)
        print((current_score, current_std))

        logfile.write(directory + "\n")
        logfile.write(str((current_score, current_std)) + "\n")
        scores.append((current_score, current_std))
        logfile.write(" -- \n\n")
    except:
        pass
    
    print(' -- ')
    print(' ')
```
