```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FID = pd.read_csv("FID.csv")
```

```python
sorted_FID = FID.sort_values("FID")
sorted_FID['Combined'] = sorted_FID['images1'] + " - " + sorted_FID['images2']
```

```python
sns.set(rc={'figure.figsize':(14.7,10.27)})

g = sns.barplot(data=sorted_FID[~(sorted_FID["images2"].str.contains("scaled"))], x='FID', y='Name')
```

```python
Inception = pd.read_csv("InceptionScores.csv")
sorted_Inception = Inception.sort_values("Inception")
```

```python
sns.set(rc={'figure.figsize':(14.7,10.27)})

g = sns.barplot(data=sorted_Inception[sorted_Inception['Name'].str.contains("GANGogh")], x='Inception', y='Name')
```

```python
sns.set(rc={'figure.figsize':(14.7,10.27)})

g = sns.barplot(data=sorted_Inception, x='Inception', y='Name', ci="Std")
```
